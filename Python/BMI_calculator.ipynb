```python
name = input("Enter your name:")
```

```python
weight = float(input("Enter your weight (kg) :"))
height = float(input("Enter your height (m) :"))
```

```python
bmi = weight / (height ** 2)
```

```python
print("BMI",round(bmi,2))
```

```python
if bmi > 0:
    if bmi < 18.5:
        print(name+" " + "Category: Underweight.")
    elif bmi < 25:
        print(name+" " + "Category: Normal weight.")
    elif bmi < 30:
        print(name +" "+ "Category: Normal weight.")    
    else:
        print(name+" " + "Category: Obesity.")
else:
    print("Not a valid value.")
```
