```python
URL ='https://www.amazon.in/ADRO-Believe-Design-Printed-T-Shirt/dp/B09NMGCZF4?ref_=ast_sto_dp&th=1&psc=1'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1"
}
page = requests.get(URL,headers=headers)
soup1 = BeautifulSoup(page.content,'html.parser')
soup2 = BeautifulSoup(soup1.prettify(),'html.parser')
title = soup2.find(id="productTitle").get_text()
price = soup2.find(class_="a-price-whole").get_text()
rating = soup2.find(class_="a-icon-alt").get_text()
title = title.strip()
price = price.strip()
rating = rating.strip()
import datetime
today = datetime.date.today()

```

```python
import csv
header = ['Title','Price','Rating','Date']
data = [title,price,rating,today]

file_path =r"C:\Users\sayan\OneDrive\Desktop\data analyst\python_practice/AmazonScapperDataset.csv"
with open(file_path,'w',newline='',encoding='UTF8')as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
print("File saved successfully!")    
```
